```python
# import required libraries
import pandas as pd
import numpy as np
```

```python
# Load data
product_data = pd.read_csv('/content/product_data.csv')
product_data
```

```python
# Dataset information
product_data.info()
```

```python
# Checking null values
null_data = product_data.isnull().sum()
null_data
```

```python
# Replace null values in 'actual_price' with corresponding values from 'price'
product_data['actual_price'] = product_data['actual_price'].fillna(product_data['price'])
product_data
```

```python
# Retrieve rows where 'price' is null
null_values = product_data[product_data['price'].isnull()]
null_values
```

```python
# Drop the row at index 336
product_data.drop(index=336, inplace=True)

# Reset index
product_data.reset_index(drop=True, inplace=True)

product_data
```

```python
# Function to clean price columns
def clean_price(data):
    try:
        data['price'] = data['price'].str.replace('Rs. ', '').str.replace(',', '').astype(float)
        data['actual_price'] = data['actual_price'].str.replace('Rs. ', '').str.replace(',', '').astype(float)
        print("Price columns cleaned!")
        return data
    except Exception as e:
        print(f"Error cleaning price columns: {e}")
        return data
```

```python
# Function to clean ratings
def clean_ratings(data):
    try:
        data['ratings'] = data['ratings'].str.extract('(\d+)').astype(float)
        print("Ratings cleaned!")
        return data
    except Exception as e:
        print(f"Error cleaning ratings: {e}")
        return data
```

```python
# Function to save cleaned data
def save_cleaned_data(data, file_path):
    try:
        data.to_csv(file_path, index=False)
        print(f"Cleaned data saved to {file_path}")
    except Exception as e:
        print(f"Error saving data: {e}")
```

```python
if product_data is not None:
    product_data = clean_price(product_data)
    product_data = clean_ratings(product_data)
    save_cleaned_data(product_data, 'cleaned_product_data.csv')
```
