```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load preprocessed data
preprocessed_product_data = pd.read_csv('/content/preprocessed_product_data.csv')
preprocessed_product_data
```

```python
# Function to plot most frequent words in Model

from collections import Counter

def visualize_frequent_words(data, column, top_n=20):
    try:
        # Tokenize words
        all_words = ' '.join(data[column]).split()
        word_counts = Counter(all_words)

        # Get most common words
        common_words = word_counts.most_common(top_n)
        words, counts = zip(*common_words)

        # Plot
        plt.figure(figsize=(10, 6))
        sns.barplot(x=counts, y=words)
        plt.title(f'Top {top_n} Most Frequent Words in {column}')
        plt.xlabel('Frequency')
        plt.ylabel('Words')
        plt.show()

    except Exception as e:
        print(f"Error in visualizing frequent words: {e}")
```

```python
# Function to visualize price vs. ratings
def visualize_price_vs_ratings(data):
    try:
        plt.figure(figsize=(10, 6))
        sns.scatterplot(x='price', y='ratings', data=data)
        plt.title('Price vs. Ratings Distribution')
        plt.xlabel('Price (Rs.)')
        plt.ylabel('Ratings')
        plt.show()

    except Exception as e:
        print(f"Error in visualizing price vs. ratings: {e}")
```

```python
# Function to create a word cloud of Model
from wordcloud import WordCloud

def visualize_wordcloud(data, column):
    try:
        # Generate the word cloud
        wordcloud = WordCloud(width=800, height=400, background_color='white').generate(' '.join(data[column]))

        # Plot the word cloud
        plt.figure(figsize=(10, 6))
        plt.imshow(wordcloud, interpolation='bilinear')
        plt.axis('off')
        plt.title(f'Word Cloud for {column}')
        plt.show()

    except Exception as e:
        print(f"Error in generating word cloud: {e}")
```

```python
# Function to visualize product distribution by ratings
def visualize_product_distribution_by_ratings(data):
    try:
        plt.figure(figsize=(10, 6))
        sns.countplot(x='ratings', data=data)
        plt.title('Product Distribution by Ratings')
        plt.xlabel('Ratings')
        plt.ylabel('Number of Products')
        plt.show()

    except Exception as e:
        print(f"Error in visualizing product distribution by ratings: {e}")
```

```python
# Function to visualize price distribution across ratings
def visualize_price_boxplot_by_ratings(data):
    try:
        plt.figure(figsize=(10, 6))
        sns.boxplot(x='ratings', y='price', data=data)
        plt.title('Price Distribution by Ratings')
        plt.xlabel('Ratings')
        plt.ylabel('Price (Rs.)')
        plt.show()

    except Exception as e:
        print(f"Error in visualizing price boxplot by ratings: {e}")
```

```python
if preprocessed_product_data is not None:
    # Visualize most frequent words in Model
    visualize_frequent_words(preprocessed_product_data, 'Model')

    # Visualize price vs. ratings
    visualize_price_vs_ratings(preprocessed_product_data)

    # Visualize word cloud of Model
    visualize_wordcloud(preprocessed_product_data, 'Model')

    # Visualize product distribution by ratings
    visualize_product_distribution_by_ratings(preprocessed_product_data)

    # Visualize price distribution across ratings
    visualize_price_boxplot_by_ratings(preprocessed_product_data)
else:
    print("Data not found. Please check the file path.")
```
